# ALL OF THE DATASET (OCT 18 - NOV 3)

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/ESP32_TEMP_COLLECTION.xlsx'

data = pd.read_excel(file_path)
data['TIME AND DATE'] = pd.to_datetime(data['TIME AND DATE'], format='%m/%d/%Y_%H:%M:%S')

# scatter plot for ESP32_TEMP_COLLECTION
plt.figure(figsize=(12, 10))
plt.subplot(2, 1, 1)
plt.scatter(data['TIME AND DATE'], data['TEMP'], marker='o', label='Temperature', color='red')
plt.title('Temperature Readings')
plt.xlabel('Time and Date')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# frequency analysis
plt.figure(figsize=(12, 6))
temperature_values = data['TEMP']
temperature_frequency = temperature_values.value_counts().sort_index()
temperature_frequency.plot(kind='bar', color='skyblue')
plt.title('Frequency Analysis of Temperature Readings')
plt.xlabel('Temperature')
plt.ylabel('Frequency')

# amplitude
amplitude = temperature_values.max() - temperature_values.min()
print(f"Amplitude: {amplitude:.2f}")

# statistics
mean_temperature = temperature_values.mean()
median_temperature = temperature_values.median()
std_deviation = temperature_values.std()

print(f"Mean Temperature: {mean_temperature:.2f}")
print(f"Median Temperature: {median_temperature}")
print(f"Standard Deviation: {std_deviation:.2f}")

plt.tight_layout()
plt.show()

```
